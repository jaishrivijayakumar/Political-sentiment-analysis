```python
# 1.Load Data

df = pd.read_csv('/content/Political Sentiment Analysis.csv') # Replace with your actual CSV file path
print("Original Data Shape:", df.shape)
print(df.head())
```

```python
# 2.Basic Info

print("\nData Types:\n", df.dtypes)
print("\nMissing Values:\n", df.isnull().sum())
```

```python
# 3.Handle Missing Values

# Separate numerical and categorical columns
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
cat_cols = df.select_dtypes(include=['object']).columns

# Numerical: Replace with mean
num_imputer = SimpleImputer(strategy='mean')
df[num_cols] = num_imputer.fit_transform(df[num_cols])

# Categorical: Replace with mode
cat_imputer = SimpleImputer(strategy='most_frequent')
df[cat_cols] = cat_imputer.fit_transform(df[cat_cols])
```

```python
# 4.Encode Categorical Variables

# Identify columns to one-hot encode based on expected features for X and y
ohe_cols = ['candidate', 'party', 'sentiment']

# Drop columns not intended for use in X or Y that are still in object format
# (e.g., identifiers like tweet_id or text fields)
df = df.drop(columns=['tweet_id', 'user_handle', 'timestamp', 'tweet_text'], errors='ignore')

# Perform One-hot Encoding for the specified categorical columns
# drop_first=True prevents multicollinearity by dropping one of the dummy variables.
df = pd.get_dummies(df, columns=ohe_cols, drop_first=True)

# No separate Label Encoding needed for binary categoricals if they are part of ohe_cols
```

```python
# 5.Outlier Detection and Treatment (using IQR)

for col in num_cols:
  Q1 = df[col].quantile(0.25)
  Q3 = df[col].quantile(0.75)
  IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
df[col] = np.where(df[col] < lower, lower, df[col])
df[col] = np.where(df[col] > upper, upper, df[col])
```

```python
# 6.Feature Scaling

scaler = StandardScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])
```

```python
import numpy as np
import sklearn.preprocessing as preprocessing
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

X = np.asarray(df[['retweets', 'likes', 'candidate_Donald Trump', 'candidate_Jill Stein','candidate_Kamala Harris', 'candidate_Robert Kennedy', 'party_Green Party', 'party_Independent', 'party_Libertarian Party', 'party_Republican Party']])
y = np.asarray(df['sentiment_positive'])

# Scale the features using preprocessing.StandardScaler()
X = preprocessing.StandardScaler().fit(X).transform(X)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
X, y, test_size = 0.3, random_state = 4)

print ('Train set:', X_train.shape, y_train.shape)
print ('Test set:', X_test.shape, y_test.shape)
```

```python
#Classification Models :

#1.Logistic Regression
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print("Accuracy:", model.score(X_test, y_test))
```

```python
#2.K-Nearest Neighbors
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=5)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print("Accuracy:", model.score(X_test, y_test))
```

```python
#3.Decision Tree
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print("Accuracy:", model.score(X_test, y_test))
```

```python
#4.Random Forest
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=100)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print("Accuracy:", model.score(X_test, y_test))
```

```python
#5.Support Vector Machine
from sklearn.svm import SVC
model = SVC()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print("Accuracy:", model.score(X_test, y_test))
```

```python
#6.Gradient Boosting
from sklearn.ensemble import GradientBoostingClassifier
model = GradientBoostingClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print("Accuracy:", model.score(X_test, y_test))
```
